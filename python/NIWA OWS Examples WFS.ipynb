# Using NIWA Open Web Service WFS feeds from Python

This example demonstrates connecting to a NIWA Web Feature Service from a Python Script.

We use the EI Browser to obtain a Filter and then show information about the resulting data.

OWSLib is a Python package for client programming with Open Geospatial Consortium (OGC) web service (hence OWS) interface standards, and their related content models.

http://geopython.github.io/OWSLib/

    easy_install OWSLib
    
see http://nbviewer.ipython.org/github/rsignell-usgs/notebook/blob/master/wms_sample.ipynb for more examples    

```python
# NIWA WMS and WFS Server endpoints are at
niwa_ows = "http://gs.niwa.co.nz/geoserver/stations/ows"
```

## NIWA Stations Catalogue WFS

### Import the WFS library and open the connection.

```python
%matplotlib inline
from owslib.wfs import WebFeatureService
wfs = WebFeatureService(niwa_ows, version='2.0.0')
```

### Display service metadata

```python
print "Title:\t", wfs.identification.title
print "Version:\t", wfs.identification.version
print "Abstract:\t", wfs.identification.abstract
print "Keywords:"
for key in wfs.identification.keywords: print "\t", key

# list operations
print "Operations:"
for op in wfs.operations: print "\t", op.name
```

### List the available layers:

```python
#Listing all available layers...
layers = list(wfs.contents)
for l in layers:  print l
```

### Details of a layer:

```python
def printlayer(layer) :
    print "Title\t", layer.title
    print "BBOX\t", layer.boundingBox
    print


printlayer(wfs.contents['stations:stations_360'])
# printlayer(wfs.contents['nemo:fbis_sampling_events'])    
```

### Getting the basic information we need to perform a GetFeature request

* All the information clients need is available in the capabilities document, which is stored in the WebFeatureService object.
* We need to choose our layer, bounding box, spatial reference system (SRS), size and format of the image.

```python
print "Operations\t", [op.name for op in wfs.operations]
print "GetFeature Methods\t", wfs.getOperationByName('GetFeature').methods
print "GetFeature Formats\t", wfs.getOperationByName('GetFeature').formatOptions
```

## GetFeature Function

### That’s everything needed to make a request for some features:

```python
stations_id = 'stations:stations_360'

# mark out a bounding box - Whole of NZ.

llong = 160
rlong = 180
tlat = -30
blat = -50

get_stations = wfs.getfeature( 
    typename=[stations_id],
    bbox=(llong, blat, rlong, tlat),
    maxfeatures = 1,
    outputFormat="application/json"
    )

# the result returned is a stream on the request so we have to read or getValue it to get the actual data.
s = get_stations.read()
```

As an aside we can use the underlying wfs methods to generate the WFS GET Request. 

```python
stations_req = wfs.getGETGetFeatureRequest( 
    typename=[stations_id],
    bbox=(llong, blat, rlong, tlat),
    maxfeatures = 1,
    outputFormat="application/json"
    )

from IPython.display import HTML
HTML("<a href='%s'>WFS GET: %s</a>" % (stations_req, stations_id ))
```

```python
import json
stations = json.loads(s)
# print json.dumps(stations, sort_keys=True,indent=4 )
# this is large so don't dump it just yet.

print "Result is a %s with %s features." % (stations['type'], stations['totalFeatures'])
print "Top Level items:", [ str(key) for key in stations]
```

```python
f = stations['features']
print "We actually received %s features." % (len(f))
print "Feature Level items:", [ str(key) for key in f[0]]


def print_feature(f):
    print "Location:\t", f['geometry']['coordinates']
    print "Geometry Name:\t", f['geometry_name']
    print "Properties:"
    for key in f['properties']: 
            print "\t", key, "\t", f['properties'][key]

for ff in f[0:2]: print_feature(ff)
# print json.dumps(f[0])
```

## Summarise the Station more decoratively using HTML

```python
from IPython.display import display, HTML

def feature_summary(f):
    display( HTML("<h2>%s <small>%s</small></h2><dl>"% (f['properties']['name'],f['properties']['station_number'] )))
    display( HTML("<dt>Location</dt><dd>%s</dd></dl>" %  f['geometry']['coordinates']))
    display( HTML("<dt>Dates</dt><dd>%s %s</dd></dl>" %  (f['properties']['start_date'], f['properties']['end_date']) ))
    display( HTML("<dt>Organisation</dt><dd>%s</dd></dl>" %  (f['properties']['organisation']) ))
    display( HTML("<dt>Timeseries Type</dt><dd>%s</dd></dl>" %  f['properties']['timeseries_type']))
    display( HTML("</dl>" ))
            
for ff in f[0:4]: feature_summary(ff)    
```

## Keyword Filter

When we use a FES filter then we omit the bbox parameter and put that into the filter.

```python
# FES is the Filter Encoding Standard.
from owslib.etree import etree
from owslib import fes
                                                                                  
# select climate stations
climate_stations = fes.PropertyIsLike('types', '*Climate*')
e =  climate_stations.toXML()
print type(e), repr(e)

e.tag
etree.tostring(e)

fr = fes.FilterRequest()
NIWA_climate_stations_xml = fr.setConstraint(climate_stations)
etree.tostring(NIWA_climate_stations_xml)
```

```python
# cut and paste filter from EI Browser.
# NIWA_climate_stations_xml = '<ogc:Filter xmlns:ogc="http://www.opengis.net/ogc"><ogc:And><ogc:And><ogc:PropertyIsLike matchCase="false" wildCard="*" singleChar="." escapeChar="!"><ogc:PropertyName>types</ogc:PropertyName><ogc:Literal>*Climate*</ogc:Literal></ogc:PropertyIsLike><ogc:PropertyIsLike matchCase="false" wildCard="*" singleChar="." escapeChar="!"><ogc:PropertyName>organisation</ogc:PropertyName><ogc:Literal>*National Institute of Water and Atmospheric Research*</ogc:Literal></ogc:PropertyIsLike></ogc:And><ogc:BBOX><ogc:PropertyName>geom</ogc:PropertyName><gml:Envelope xmlns:gml="http://www.opengis.net/gml"><gml:lowerCorner>174.65993004608075 -36.96829525677883</gml:lowerCorner><gml:upperCorner>175.065394218443 -36.715242377432304</gml:upperCorner></gml:Envelope></ogc:BBOX></ogc:And></ogc:Filter>'

req = wfs.getGETGetFeatureRequest(
    typename=[stations_id],
    filter= etree.tostring(NIWA_climate_stations_xml),
    outputFormat="application/json"
    )

req
#from owslib.etree import etree
# NIWA_climate_stations_fes = etree.fromstring(NIWA_climate_stations_xml)
```

```python
get_stations = wfs.getfeature( 
    typename=[stations_id],
    filter= NIWA_climate_stations_xml,
    outputFormat="application/json"
    )

# the result returned is a stream on the request so we have to read or getValue it to get the actual data.
stations = json.loads(get_stations.read())
print "Result is a %s with %s features." % (stations['type'], stations['totalFeatures'])
for ff in stations['features'][0:4]: feature_summary(ff)  
```
